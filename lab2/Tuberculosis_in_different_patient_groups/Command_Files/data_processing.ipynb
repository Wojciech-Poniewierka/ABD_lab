Uporządkuję bazę danych o gruźlicy w różnych grupach pacjentów.

```python
import pandas as pd
import numpy as np

data = pd.read_csv("../Original_Data/tb.csv")

PREFIX_LEN = len("new_sp_")
DATA_LEN = len(data)

# dodatkowa funkcja, która przyjmuje nazwę kolumny i zwraca płeć oraz zakres wiekowy
def extract_info(column_name):
    sex = column_name[PREFIX_LEN]
    years = column_name[PREFIX_LEN+1:]
    if len(years) == 2 and years[0] == '0':
        years = years[0] + '--' + years[1]
    elif len(years) == 2 and years[0] != '0':
        years = years + '+'
    elif len(years) == 3:
        years = years[0] + '--' + years[1:]
    elif len(years) == 4:
        years = years[:2] + '--' + years[2:]
    
    return sex, years

data.head()
```

```python
new_df = {}
new_df['iso2'] = []
new_df['year'] = []
new_df['sex'] = []
new_df['age'] = []
new_df['cases'] = []

for i in range(DATA_LEN):
    iso = data['iso2'][i]    # uzyskuję informację o iso
    year = data['year'][i]   # uzyskuję informację o roku
    for idx, col in enumerate(data.columns):
        if idx in (0, 1, 2): # pomijam pierwsze trzy kolumny
            continue
        sex, age = extract_info(col)    # uzyskuję informacje o wieku i płci
        if pd.isnull(data.loc[i, col]): # pomijam puste komórki
            continue 
        cases = data[col][i]            # uzyskuję informacje o liczbie przypadków
        # wpisuje uzyskane dane do słownika, który później przekształcę w DataFrame
        new_df['iso2'].append(iso)
        new_df['year'].append(year)
        new_df['sex'].append(sex)
        new_df['age'].append(age)
        new_df['cases'].append(int(cases))

# przekształcam słownik w nowy DataFrame
df_processed = pd.DataFrame(new_df)
```
