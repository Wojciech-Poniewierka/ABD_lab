*Zadanie 3*

funkcja f(x) realizuje funkcję $f(x) = x^2 + 5$, a funkcja plot przyjmuje wektor v, oblicza wektor wartości f(v) a następnie rysuje wykres z otrzymanych danych. 

```python
def f(x):
    return x**2 + 5


def plot(v):
    f_v = f(v)
    plt.plot(v, f_v)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title("y = f(x)")
    plt.grid()
    plt.legend("y")
```

```python
v1 = np.linspace(-1, 1, 100)
plot(v1)
```

```python
v2 = np.linspace(-6, 6, 1000)
plot(v2)
```

```python
v3 = np.linspace(0, 5, 1000)
plot(v3)
```
